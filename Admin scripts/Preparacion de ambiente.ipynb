# External location

#### creacion de external location administracion


```python
%sql
CREATE EXTERNAL LOCATION IF NOT EXISTS `extl-administracion`
URL 'abfss://administracion@${storageName}.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL credential)
```
